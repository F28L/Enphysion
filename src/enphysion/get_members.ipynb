```python
def get_all_club_members() -> dict:
    from secrets import CLUB_ID, VIRTUAGYM_API_KEY, VIRTUAGYM_CLUB_SECRET
    import requests
    from data_structures.user import User 
    url = f"https://api.virtuagym.com/api/v1/club/{CLUB_ID}/member?api_key={VIRTUAGYM_API_KEY}&club_secret={VIRTUAGYM_CLUB_SECRET}&sync_from=0"
    ret = requests.get(url)

    users = {}

    for item in ret:
        print
        user = User(
            item["member_id"],
            item["firstname"],
            item["lastname"],
            item["active"],
            item["is_pro"],
            item["gender"],
            item["email"],
            item["member_since"],
            item["timestamp_edit"],
            item["country"],
            item["club_id"],
            item["registration_date"],
            item["lang"],
            item["original_member_id"],
            item.get("birthday")  # Check if "birthday" exists
        )
        users[item["member_id"]] = user.to_dict()  # Convert User object to dictionary

    return users

```
